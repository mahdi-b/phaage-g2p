```python
import umap
import umap.plot
import numpy as np
```

```python
import sklearn.datasets
```

```python
pendigits = sklearn.datasets.load_digits()
```

```python
pendigits.data.shape
```

```python
mapper = umap.UMAP().fit(pendigits.data)
```

```python
umap.plot.points(mapper, labels=pendigits.target)
```

```python
embeddings =  []
data = {"ids":[], "targets":[]}
```

```python
for line in open("2337_representation.tsv"):
    line_data = line.split("\t")
    data["ids"].append(line_data[0])
    embeddings.append(np.array(eval(line_data[1])))
    data["targets"].append("2337")
```

```python
for line in open("2478_representation.tsv"):
    line_data = line.split("\t")
    data["ids"].append(line_data[0])
    embeddings.append(np.array(eval(line_data[1])))
    data["targets"].append("2478")
```

```python
for line in open("random.tsv"):
    line_data = line.split("\t")
    data["ids"].append(line_data[0])
    embeddings.append(np.array(eval(line_data[1])))
    data["targets"].append("random")
```

```python
data["embeddings"] = np.array(embeddings)
```

```python
data["embeddings"].shape
```

```python
mapper = umap.UMAP().fit(data["embeddings"])
```

```python
umap.plot.points(mapper, labels=np.array(data["targets"]))
```

```python
from scipy.spatial import distance
import itertools
```

```python
distnace_clust_1 = []
for comb in list(itertools.combinations(range(60),2)):
    distnace_clust_1.append(distance.cosine(data["embeddings"][comb[0]], data["embeddings"][comb[1]]))
min(distnace_clust_1), max(distnace_clust_1)
```

```python
distnace_clust_2 = []
for comb in list(itertools.combinations(range(61,90),2)):
    distnace_clust_2.append(distance.cosine(data["embeddings"][comb[0]], data["embeddings"][comb[1]]))
min(distnace_clust_2), max(distnace_clust_2)
```

```python
distnace_clust_3 = []
for comb in list(itertools.combinations(range(90, 120),2)):
    distnace_clust_3.append(distance.cosine(data["embeddings"][comb[0]], data["embeddings"][comb[1]]))
min(distnace_clust_3), max(distnace_clust_3)
```

```python
distnace_clust_1_2 = []
for comb in list(list(itertools.product(range(60), range(61,90)))):
    distnace_clust_1_2.append(distance.cosine(data["embeddings"][comb[0]], data["embeddings"][comb[1]]))
min(distnace_clust_1_2), max(distnace_clust_1_2)

```
